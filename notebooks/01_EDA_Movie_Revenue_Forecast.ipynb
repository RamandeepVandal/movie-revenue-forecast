# Movies Revenues Forecast: EDA

```python
# Import libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv
from urllib.parse import quote_plus
plt.style.use("ggplot")
```

```python
# Connect to DB
load_dotenv()

password = quote_plus(os.getenv("DB_PASS"))
dbname = os.getenv("DB_NAME")

connection_str = f"postgresql+psycopg2://postgres:{password}@{dbname}"
engine = create_engine(connection_str)
```

```python
# Load in data to df
movie_df = pd.read_sql('SELECT * FROM movies_cleaned', engine)
```

```python
# convert release_date column to datetime
movie_df['release_date'] = pd.to_datetime(movie_df['release_date'], errors='coerce')
```

```python
# update the column with only years
movie_df['release_date'] = movie_df['release_date'].dt.year
```

Missing Data Handling 

```python
# check for nulls
movie_df.isnull().sum()
```

```python
# drop null values
movie_df.dropna()
```

```python
# check for duplicates
movie_df.duplicated().sum()
```

## Cleaning

```python
# Clean genre column
```

```python
movie_df['genres'] = movie_df['genres'].apply(lambda x: x.split(',')[0] if pd.notnull(x) else x)
```

```python
# Clean procution companies column
```

```python
movie_df['production_companies_list'] = movie_df['production_companies_list'].apply(lambda x: x.split(',')[0] if pd.notnull(x) else x)
```

```python
# Clean production countries column 
```

```python
movie_df['production_countries_list'] = movie_df['production_countries_list'].apply(lambda x: x.split(',')[0] if pd.notnull(x) else x)
```

## Questions


```python
# 1. Which genres earn the most revenue
```

```python
revenue_by_genre = movie_df.groupby('genres')['revenue'].sum().sort_values(ascending=False)
```

```python
revenue_by_genre.plot(kind='barh')
```

```python
# 2. Which genres cost the most to create
```

```python
budget_by_genre = movie_df.groupby('genres')['budget'].sum().sort_values(ascending=False)
```

```python
budget_by_genre.plot(kind='barh')
```

```python
# 3. Top 10 most profitable companies -> create new column -> profit
```

```python
revenue_by_company = movie_df.groupby('production_companies_list')['revenue'].sum().sort_values(ascending=False).head(10)
revenue_by_company.plot(kind='barh')
```

```python
# 4. Impact on budget based on runtime
```

```python
movie_df.plot(kind='scatter', x='runtime', y='budget', figsize=(18, 12), alpha=0.5)
```

```python
# 5. Ratings based on release date
```

```python
movie_df.plot(kind='scatter', x='release_date', y='vote_average', figsize=(18, 12))
```

```python
# 6. Impact on budget, revenue and profit based on production countries
```

```python
country_grouped = movie_df.groupby('production_countries_list')[['budget', 'revenue']].sum()
top_countries = country_grouped.sort_values('budget', ascending=True).tail(8)
top_countries.plot(kind='line', figsize=(18, 12), marker='o')
```
